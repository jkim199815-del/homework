```python
import pandas as pd
```

```python
df = pd.read_csv('games.csv') 
```

```python
genre_na_sales = df.groupby('Genre')['NA_sales'].sum()
print(genre_na_sales)
```

```python
platform_mean_sales = df.groupby('Platform')[['NA_sales', 'EU_sales','JP_sales', 'Other_sales']].mean()
print(platform_mean_sales)
```

```python
extra_ratings = pd.DataFrame({'Name': ['Wii Sports', 'Super Mario Bros.', 'Mario Kart Wii'],'Critic_Score': [80, 90, 85]})
```

```python
extra_ratings = pd.DataFrame({'Name': ['Wii Sports', 'Super Mario Bros.', 'Mario Kart Wii'],'Critic_Score': [80, 90, 85]})

merged_df = df.merge(extra_ratings, on='Name', how='left')
merged_df.head()
```

```python
pivot_na_eu = df.pivot_table(index='Platform',columns='Genre',values=['NA_sales', 'EU_sales'],aggfunc='mean',fill_value=0)

pivot_na_eu
```

```python
platform_genre_jp_mean = df.groupby(['Platform', 'Genre'])['JP_sales'].mean()
platform_genre_jp_mean
```

```python
platform_genre_jp_mean = platform_genre_jp_mean.unstack(fill_value=0)
platform_genre_jp_mean
```

```python
pivot_na_max = df.pivot_table(index='Platform',columns='Genre',values='NA_sales',aggfunc='max',
    fill_value=0
)

pivot_na_max
```
